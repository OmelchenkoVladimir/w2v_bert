```python
import ast
import numpy as np
import pandas as pd
import gensim

from tensorboardX import SummaryWriter
```

```python
train_sample = pd.read_csv('rus/bert/train_sample.csv')
valid_sample = pd.read_csv('rus/bert/valid_sample.csv')
```

```python
w2v = gensim.models.KeyedVectors.load_word2vec_format('w2v_models/all_norm-sz500-w10-cb0-it3-min5.w2v', binary=True, unicode_errors='ignore')
```

```python
w2v.init_sims(replace=True)
```

```python
train_sample['closest_embedding'] = train_sample['embedding'].apply(lambda x: w2v.most_similar([np.array(ast.literal_eval(x))])[1][0])
valid_sample['closest_embedding'] = valid_sample['embedding'].apply(lambda x: w2v.most_similar([np.array(ast.literal_eval(x))])[1][0])
```

```python
def find_pos(row, epoch):
    lst = w2v.most_similar([np.array(ast.literal_eval(row[f'emb_epoch{epoch}']))], topn=len(w2v.vocab)+1)
    words = [x for (x,y) in lst]
    return words.index(row['closest_embedding'])
```

```python
for i in range(1, 21):
    print(i)
    train_sample[f'most_similar_{i}'] = train_sample[f'emb_epoch{i}'].apply(lambda x: w2v.most_similar([np.array(ast.literal_eval(x))], topn=25))
    valid_sample[f'most_similar_{i}'] = valid_sample[f'emb_epoch{i}'].apply(lambda x: w2v.most_similar([np.array(ast.literal_eval(x))], topn=25))
    train_sample[f'closest_pos_{i}'] = train_sample.apply(lambda x: find_pos(x, i), axis=1)
    valid_sample[f'closest_pos_{i}'] = valid_sample.apply(lambda x: find_pos(x, i), axis=1)
```

```python
writer = SummaryWriter()

for i in range(1, 21):
    writer.add_scalars('Metrics', {'train_closest':train_sample[f'closest_pos_{i}'].median(), 'valid_closest':valid_sample[f'closest_pos_{i}'].median()}, i)

writer.close()
```

```python
train_sample.to_csv('rus/bert/train_sample_metrics.csv', index=None)
valid_sample.to_csv('rus/bert/valid_sample_metrics.csv', index=None)
```
