```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('freqrnc2011.csv', sep = '\t')
```

```python
data.sort_values(by='Freq(ipm)', ascending=False)
```

```python
hfreq = data[data['Freq(ipm)'] > 300]
```

```python
lfreq = data[data['Freq(ipm)'] < 10]
print(lfreq.shape)
```

```python
mfreq = data[(data['Freq(ipm)'] >= 10) & (data['Freq(ipm)'] <= 300)]
print(mfreq.shape)
```

```python
hfreq_words = hfreq['Lemma'].unique()
```

```python
mfreq_words = np.array([word for word in mfreq['Lemma'].unique() if word not in hfreq_words])
```

```python
mfreq_words
```

```python
lfreq_words = np.array([word for word in lfreq['Lemma'].unique() if ((word not in hfreq_words) and (word not in mfreq_words))])
```

```python
print(hfreq_words.shape)
print(mfreq_words.shape)
print(lfreq_words.shape)
```

```python
np.save('hfreq_words.npy', hfreq_words)
np.save('mfreq_words.npy', mfreq_words)
np.save('lfreq_words.npy', lfreq_words)
```
