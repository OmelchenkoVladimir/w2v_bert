```python
import pandas as pd
```

```python
simlex = pd.read_csv('SimLex-999/SimLex-999.txt', sep = '\t')[['word1', 'word2', 'SimLex999']]
```

```python
model = gensim.models.KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True)
```

```python
model
```

```python
simlex['w2v'] = simlex.apply(lambda x: model.similarity(x['word1'], x['word2']), axis = 1)
```

```python
from scipy.stats import spearmanr
```

```python
spearmanr(simlex['SimLex999'], simlex['w2v'])
```

```python
rw = pd.read_csv('rw/rw.txt', sep = '\t', header = None)[[0, 1, 2]].rename({0:'word1', 1:'word2', 2:'rw'}, axis = 1)
```

```python
def get_word_similarity(model, row, default_value):
    try:
        return model.similarity(row['word1'], row['word2'])
    except KeyError:
        return default_value
```

```python
rw['w2v'] = rw.apply(lambda x: get_word_similarity(model, x, 0), axis = 1)
```

```python
rw
```

```python
spearmanr(rw['rw'], rw['w2v'])
```
