```python
import numpy as np
import pandas as pd

import gensim
```

```python
w2v = gensim.models.KeyedVectors.load_word2vec_format('w2v_models/all_norm-sz500-w10-cb0-it3-min5.w2v', binary=True, unicode_errors='ignore')
```

```python
test = pd.read_csv('russe-evaluation/russe/evaluation/test.csv')
```

```python
def w2v_distance(w2v, word1, word2):
    try:
        return w2v.distance(word1, word2)
    except KeyError:
        return np.nan
```

```python
test['sim'] = test.apply(lambda row: 1 - w2v_distance(w2v, row['word1'], row['word2']), axis = 1)
```

```python
w2v.vocab
```

```python
test['sim'] = test['sim'].fillna(test['sim'].mean())
```

```python
test.to_csv('russe-evaluation/russe/evaluation/SUM_gallw2v_withnan.csv', index=None) # 0.69920, 0.91870, 0.94632, 0.97732
```
